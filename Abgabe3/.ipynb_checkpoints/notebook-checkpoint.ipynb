```python
import ipyleaflet
from ipywidgets import HTML
from ipyleaflet import Map, Marker, Popup
import pandas as pd
import geocoder
import ipywidgets as widgets
from ipywidgets import interact


airlines = []
center =(50,10)
        
flightsAMS = pd.read_csv('aerodata_flughafen_amsterdam.csv', sep=';')

flightsVIE = pd.read_csv('aerodata_flughafen_wien.csv', sep=';')
flightsFRA = pd.read_csv('aerodata_flughafen_frankfurt.csv', sep=';')
delayedAMS = 0
sumAMS = 0
i = 0
while i<len(flightsAMS):
    if flightsAMS['airline'][i] not in airlines:
        if "{" not in flightsAMS['airline'][i]:
            airlines.append(flightsAMS['airline'][i])
    if(flightsAMS['status'][i]=='Delayed'):
        sumAMS = sumAMS + (flightsAMS['actualTime'][i]*1000000000-flightsAMS['scheduledTime'][i]*1000000000)
        delayedAMS=delayedAMS+1
    i=i+1

    
delayAMS = pd.to_timedelta(sumAMS.astype('timedelta64[ns]'))

delayedVIE = 0
sumVIE= 0
i=0
while i<len(flightsVIE):
    if flightsVIE['airline'][i] not in airlines:
        if "{" not in flightsVIE['airline'][i]:
            airlines.append(flightsVIE['airline'][i])
    if(flightsVIE['status'][i]=='Delayed'):
        sumVIE = sumVIE + (flightsVIE['actualTime'][i]*1000000000-flightsVIE['scheduledTime'][i]*1000000000)
        delayedVIE=delayedVIE+1
    i=i+1
    
delayVIE = pd.to_timedelta(sumVIE.astype('timedelta64[ns]'))

delayedFRA = 0
sumFRA= 0
i=0
while i<len(flightsFRA):
    if flightsFRA['airline'][i] not in airlines:
        if "{" not in flightsFRA['airline'][i]:
            airlines.append(flightsFRA['airline'][i])
    if(flightsFRA['status'][i]=='Delayed'):
        sumFRA = sumFRA + (flightsFRA['actualTime'][i]*1000000000-flightsFRA['scheduledTime'][i]*1000000000)
        delayedFRA=delayedFRA+1
    i=i+1
delayFRA = pd.to_timedelta(sumFRA.astype('timedelta64[ns]'))


locationVIE = geocoder.osm('Vienna, Austria')
locationFRA = geocoder.osm('Frankfurt, Germany')
locationAMS = geocoder.osm('Amsterdam, NEtherlands')


latlngVIE = [locationVIE.lat, locationVIE.lng]
latlngFRA = [locationFRA.lat, locationFRA.lng]
latlngAMS = [locationAMS.lat, locationAMS.lng]


#vie = Map(center=latlng)
markerVIE = Marker(location=latlngVIE, title='Vienna')
markerFRA = Marker(location=latlngFRA, title='Frankfurt')
markerAMS = Marker(location=latlngAMS, title='Amsterdam')

#markerAMS.on_mouseover(test())
#vie.add_layer(marker)
basic_map = Map(center =(50,10),zoom=5)
basic_map.add_layer(markerVIE)
basic_map.add_layer(markerFRA)
basic_map.add_layer(markerAMS)

message1 = HTML()
message2 = HTML()
message3 = HTML()
message4 = HTML()

message1.value = "Um zu den Daten zu gelangen, klicken Sie auf die jeweilige Makierung!"
message2.value = str(delayAMS) +"<br>"+str(delayedAMS)

message2.placeholder = "Some HTML"
message2.description = "Avg. Delay: <br> # of delays: "

message3.value = str(delayVIE)+"<br>"+str(delayedVIE)

                                                                
message3.placeholder = "Some HTML"
message3.description = "Avg. Delay: <br> # of delays: "

message4.value = str(delayFRA)+"<br>"+str(delayedFRA)
message4.placeholder = "Some HTML"
message4.description = "Avg. Delay: <br> # of delays: "

# Popup with a given location on the map:
popup = Popup(
    location=center,
    child=message1,
    close_button=True,
    auto_close=True,
    close_on_escape_key=False
)

basic_map.add_layer(popup)

# Popup associated to a layer
markerAMS.popup = message2
markerVIE.popup = message3
markerFRA.popup = message4

    
def f(Dataset):
    control = Dataset
    if control != 'nothing selected':
        sumAMS = 0
        delayedAMS = 0
        i = 0
        while i<len(flightsAMS):    
            if(flightsAMS['status'][i]=='Delayed' and flightsAMS['airline'][i]==control):
                sumAMS = sumAMS + (flightsAMS['actualTime'][i]*1000000000-flightsAMS['scheduledTime'][i]*1000000000)
                delayedAMS=delayedAMS+1
            i=i+1
        delayAMS= 0  
        if(sumAMS != 0):
            delayAMS = pd.to_timedelta(sumAMS.astype('timedelta64[ns]'))
 
        
        sumVIE = 0
        delayedVIE = 0
        i = 0
        while i<len(flightsVIE):    
            if(flightsVIE['status'][i]=='Delayed' and flightsVIE['airline'][i]==control):
                sumVIE = sumVIE + (flightsVIE['actualTime'][i]*1000000000-flightsVIE['scheduledTime'][i]*1000000000)
                delayedVIE=delayedVIE+1
            i=i+1
            
        delayVIE = 0
        if(sumVIE!=0):
            delayVIE = pd.to_timedelta(sumVIE.astype('timedelta64[ns]'))
            
        sumFRA = 0
        delayedFRA = 0
        i = 0
        while i<len(flightsFRA):    
            if(flightsFRA['status'][i]=='Delayed' and flightsFRA['airline'][i]==control):
                sumFRA = sumFRA + (flightsFRA['actualTime'][i]*1000000000-flightsFRA['scheduledTime'][i]*1000000000)
                delayedFRA=delayedFRA+1
            i=i+1
            
        delayFRA = 0
        if(sumFRA!=0):
            delayFRA = pd.to_timedelta(sumFRA.astype('timedelta64[ns]'))

        message2.value = str(delayAMS) +"<br>"+str(delayedAMS)
        message3.value = str(delayVIE)+"<br>"+str(delayedVIE)
        message4.value = str(delayFRA)+"<br>"+str(delayedFRA)
        markerAMS.popup = message2
        markerVIE.popup = message3
        markerFRA.popup = message4
        

airlines.sort()
airlines.insert(0, 'nothing selected')

interact(f, Dataset = widgets.Dropdown(
options=airlines,
description='Airline:',
disabled=False))
basic_map
```
